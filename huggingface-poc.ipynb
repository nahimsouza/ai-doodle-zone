### English

```python
from transformers import pipeline

oracle = pipeline(model="deepset/roberta-base-squad2")

context = """
Wolfgang Amadeus Mozart[a][b] (27 January 1756 – 5 December 1791) was a prolific and influential composer of the Classical period. Despite his short life, his rapid pace of composition resulted in more than 800 works representing virtually every Western classical genre of his time. Many of these compositions are acknowledged as pinnacles of the symphonic, concertante, chamber, operatic, and choral repertoire. Mozart is widely regarded as being one of the greatest composers in the history of Western music,[1] with his music admired for its "melodic beauty, its formal elegance and its richness of harmony and texture".[2]

Born in Salzburg, then in the Holy Roman Empire and currently in Austria, Mozart showed prodigious ability from his earliest childhood. At the age of five he was already competent on keyboard and violin, he had begun to compose, and he performed before European royalty. His father took him on a grand tour of Europe and then three trips to Italy. At 17, he was a musician at the Salzburg court but grew restless and travelled in search of a better position.

While visiting Vienna in 1781, Mozart was dismissed from his Salzburg position. He stayed in Vienna, where he achieved fame but little financial security. During his final years there, he composed many of his best-known symphonies, concertos, and operas. His Requiem was largely unfinished at the time of his death at the age of 35, the circumstances of which are uncertain and much mythologised.
"""

questions = [
    "Who is Mozart?",
    "When did Mozart died?",
    "Where did Mozart died?",
    "When did Mozart visited Viena?",
    "How many musics did Mozart compose?",
    "Where Mozart was born?",
    "When Mozart was born?",
    "Did Mozart visit Brazil?",
    "Which instruments did Mozart play?"
]

for question in questions:
    print(f"Question: {question}, Ans: {oracle(question=question, context=context)}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

tokenizer = AutoTokenizer.from_pretrained('nicholasKluge/TeenyTinyLlama-460m-Chat')
model = AutoModelForCausalLM.from_pretrained('nicholasKluge/TeenyTinyLlama-460m-Chat')

model.eval()
model.to(device)

context_ptbr = """
Considere o texto a seguir para responder à pergunta.

Wolfgang Amadeus Mozart (27 de janeiro de 1756 - 5 de dezembro de 1791) foi um prolífico e influente compositor do período Clássico. Apesar de sua vida curta, seu ritmo acelerado de composição resultou em mais de 800 obras representando praticamente todos os gêneros clássicos ocidentais de sua época. Muitas dessas composições são reconhecidas como pináculos do repertório sinfônico, concertante, de câmara, operístico e coral. Mozart é amplamente considerado como um dos maiores compositores na história da música ocidental, com sua música sendo admirada por sua "beleza melódica, sua elegância formal e sua riqueza de harmonia e textura".

Nascido em Salzburgo, então no Sacro Império Romano e atualmente na Áustria, Mozart mostrou habilidade prodigiosa desde a mais tenra infância. Aos cinco anos, já era competente no teclado e no violino, já havia começado a compor e se apresentava diante da realeza europeia. Seu pai o levou em uma grande turnê pela Europa e depois em três viagens à Itália. Aos 17 anos, ele era músico na corte de Salzburgo, mas ficou inquieto e viajou em busca de uma posição melhor.

Ao visitar Viena em 1781, Mozart foi demitido de seu cargo em Salzburgo. Ele permaneceu em Viena, onde alcançou fama, mas pouco segurança financeira. Durante seus últimos anos lá, ele compôs muitas de suas sinfonias, concertos e óperas mais conhecidos. Seu Requiem estava em grande parte inacabado no momento de sua morte, aos 35 anos, cujas circunstâncias são incertas e muito mitificadas.

Pergunta: """


questions = [
    "Quem é Mozart?",
    "Quando Mozart morreu?",
    "Onde Mozart morreu?",
    "Quando Mozart visitou Viena?",
    "Quantas músicas Mozart compôs?",
    "Onde Mozart nasceu?",
    "Quando Mozart nasceu?",
    "Mozart visitou o Brasil?",
    "Quais instrumentos Mozart tocava?"
]


# question =  input("Entre seu prompt aqui: ")
for question in questions:
    inputs = tokenizer("<instruction>" + context_ptbr + question + "</instruction>", return_tensors="pt").to(device)
    responses = model.generate(**inputs, num_return_sequences=2)
    print(f"Pergunta: 👤 {question}\n")

    for i, response in  enumerate(responses):
        print(f'Resposta {i+1}: 🤖 {tokenizer.decode(response, skip_special_tokens=True).replace(context_ptbr + question, "")}')
```
