### English

```python
from transformers import pipeline

oracle = pipeline(model="deepset/roberta-base-squad2")

context = """
Wolfgang Amadeus Mozart[a][b] (27 January 1756 ‚Äì 5 December 1791) was a prolific and influential composer of the Classical period. Despite his short life, his rapid pace of composition resulted in more than 800 works representing virtually every Western classical genre of his time. Many of these compositions are acknowledged as pinnacles of the symphonic, concertante, chamber, operatic, and choral repertoire. Mozart is widely regarded as being one of the greatest composers in the history of Western music,[1] with his music admired for its "melodic beauty, its formal elegance and its richness of harmony and texture".[2]

Born in Salzburg, then in the Holy Roman Empire and currently in Austria, Mozart showed prodigious ability from his earliest childhood. At the age of five he was already competent on keyboard and violin, he had begun to compose, and he performed before European royalty. His father took him on a grand tour of Europe and then three trips to Italy. At 17, he was a musician at the Salzburg court but grew restless and travelled in search of a better position.

While visiting Vienna in 1781, Mozart was dismissed from his Salzburg position. He stayed in Vienna, where he achieved fame but little financial security. During his final years there, he composed many of his best-known symphonies, concertos, and operas. His Requiem was largely unfinished at the time of his death at the age of 35, the circumstances of which are uncertain and much mythologised.
"""

questions = [
    "Who is Mozart?",
    "When did Mozart died?",
    "Where did Mozart died?",
    "When did Mozart visited Viena?",
    "How many musics did Mozart compose?",
    "Where Mozart was born?",
    "When Mozart was born?",
    "Did Mozart visit Brazil?",
    "Which instruments did Mozart play?"
]

for question in questions:
    print(f"Question: {question}, Ans: {oracle(question=question, context=context)}")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

tokenizer = AutoTokenizer.from_pretrained('nicholasKluge/TeenyTinyLlama-460m-Chat')
model = AutoModelForCausalLM.from_pretrained('nicholasKluge/TeenyTinyLlama-460m-Chat')

model.eval()
model.to(device)

context_ptbr = """
Considere o texto a seguir para responder √† pergunta.

Wolfgang Amadeus Mozart (27 de janeiro de 1756 - 5 de dezembro de 1791) foi um prol√≠fico e influente compositor do per√≠odo Cl√°ssico. Apesar de sua vida curta, seu ritmo acelerado de composi√ß√£o resultou em mais de 800 obras representando praticamente todos os g√™neros cl√°ssicos ocidentais de sua √©poca. Muitas dessas composi√ß√µes s√£o reconhecidas como pin√°culos do repert√≥rio sinf√¥nico, concertante, de c√¢mara, oper√≠stico e coral. Mozart √© amplamente considerado como um dos maiores compositores na hist√≥ria da m√∫sica ocidental, com sua m√∫sica sendo admirada por sua "beleza mel√≥dica, sua eleg√¢ncia formal e sua riqueza de harmonia e textura".

Nascido em Salzburgo, ent√£o no Sacro Imp√©rio Romano e atualmente na √Åustria, Mozart mostrou habilidade prodigiosa desde a mais tenra inf√¢ncia. Aos cinco anos, j√° era competente no teclado e no violino, j√° havia come√ßado a compor e se apresentava diante da realeza europeia. Seu pai o levou em uma grande turn√™ pela Europa e depois em tr√™s viagens √† It√°lia. Aos 17 anos, ele era m√∫sico na corte de Salzburgo, mas ficou inquieto e viajou em busca de uma posi√ß√£o melhor.

Ao visitar Viena em 1781, Mozart foi demitido de seu cargo em Salzburgo. Ele permaneceu em Viena, onde alcan√ßou fama, mas pouco seguran√ßa financeira. Durante seus √∫ltimos anos l√°, ele comp√¥s muitas de suas sinfonias, concertos e √≥peras mais conhecidos. Seu Requiem estava em grande parte inacabado no momento de sua morte, aos 35 anos, cujas circunst√¢ncias s√£o incertas e muito mitificadas.

Pergunta: """


questions = [
    "Quem √© Mozart?",
    "Quando Mozart morreu?",
    "Onde Mozart morreu?",
    "Quando Mozart visitou Viena?",
    "Quantas m√∫sicas Mozart comp√¥s?",
    "Onde Mozart nasceu?",
    "Quando Mozart nasceu?",
    "Mozart visitou o Brasil?",
    "Quais instrumentos Mozart tocava?"
]


# question =  input("Entre seu prompt aqui: ")
for question in questions:
    inputs = tokenizer("<instruction>" + context_ptbr + question + "</instruction>", return_tensors="pt").to(device)
    responses = model.generate(**inputs, num_return_sequences=2)
    print(f"Pergunta: üë§ {question}\n")

    for i, response in  enumerate(responses):
        print(f'Resposta {i+1}: ü§ñ {tokenizer.decode(response, skip_special_tokens=True).replace(context_ptbr + question, "")}')
```
